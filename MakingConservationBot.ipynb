# Making Bot Model

```python
# Import the model tokenizer and bot generate model
from transformers import BlenderbotTokenizer, BlenderbotForConditionalGeneration
```

```python
mdata = "facebook/blenderbot-400M-distill"
```

```python
# Setup Token and  Model
tokenizer = BlenderbotTokenizer.from_pretrained(mdata)
model = BlenderbotForConditionalGeneration.from_pretrained(mdata)
```

# Conservation section

```python
UTTERANCE = "My name is Zin Paing Oo.I'm a data scientist."
```

```python
# start conservation text to token and get return result
inputs = tokenizer([UTTERANCE], return_tensors="pt")
inputs
```

```python
# passing through blender model and get result by using model
reply_ids = model.generate(**inputs)
```

```python
# Decodde the model output
tokenizer.batch_decode(reply_ids)
```

# All in one together function

```python
def conservation(TEXT_CON):
    TEXT = TEXT_CON
    inputs = tokenizer([TEXT], return_tensors="pt")
    reply_ids = model.generate(**inputs)
    return tokenizer.batch_decode(reply_ids)
```

```python
conservation('Hello , Can you explain about data scientist career?')
```

```python
conservation('Hello ,  My name is Zin Paing Oo. What is your name?')
```

```python
conservation('Nice to meet you too, I am a data analyst.')
```
